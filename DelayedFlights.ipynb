```python
missing_values=data.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python
missing_values_percentage = (data.isnull().sum() / len(data)) * 100
print("Percentage of missing values per column:")
print(missing_values_percentage)
```

```python
data.columns
```

```python
data.rename(columns={'DayofMonth': 'Day'}, inplace=True)
```

```python
data.columns
```

```python
delay_columns = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
data[delay_columns] = data[delay_columns].fillna(0)
print("Missing values after filling delay columns with 0:")
print(data[delay_columns].isnull().sum())
```

```python
data['Route'] = data['Origin'] + ' to ' + data['Dest']
display(data[['Origin', 'Dest', 'Route']].head())
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

data['Season'] = data['Month'].apply(get_season)
display(data[['Month', 'Season']].head())
```

```python
print("Shape before filtering cancelled and diverted flights:", data.shape)
data_filtered = data[(data['Cancelled'] == 0) & (data['Diverted'] == 0)].copy()
print("Shape after filtering cancelled and diverted flights:", data_filtered.shape)
```

```python
flights_per_route = data_filtered.groupby('Route').size().reset_index(name='Number of Flights')
display(flights_per_route.head())
```

```python
data_filtered.to_csv('cleaned_flights_data.csv', index=False)
print("Cleaned dataset saved to 'cleaned_flights_data.csv'")
```
